```python
## Bar plot()
brand =list(df['brand'].value_counts().keys())  ## catgorical(.keys())
no_of_counts=list(df['brand'].value_counts().values)  ## numerical(.values)
plt.figure(figsize=(16,8))
plt.bar(brand,no_of_counts,color='green',ec='r',lw=4,width=1.3)   # ec=edge color   # lw=line width  # ls=dashed,dotted,solid
plt.xlabel('BRANDS')
plt.ylabel('no_of_counts')
plt.title('No. of BIKE in Brand')
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.show()
```

```python
## line plot graph is used to check the trend
## scatter plot is used to check the realtionship or distribution between them and both columns are numerical
## bar plot is used to check the amount or the quantity
```

```python
df.drop_duplicates(inplace=True)
```

```python
bullet = df[df['brand']=='Royal Enfield']
```

```python
ownership = list(bullet['owner'].value_counts().keys())  ## catgorical
no_of_counts=list(bullet['owner'].value_counts().values)  ## numerical
```

```python
ownership=['First Owner','Second Owner','Third Owner','Fourth Owner']
```

```python
no_of_counts
```

```python
## Pieplot()
plt.figure(figsize=(8,8))
plt.pie(no_of_counts,labels=ownership,autopct='%1.2f%%')
plt.show()
```

```python
country=['India','China','Japan','Pakistan']
population=[900,800,500,350]
explosion=[0,0.2,0,0.4]
plt.figure(figsize=(8,8))
plt.pie(population,labels=country,autopct='%1.2f%%',wedgeprops={'ec':'k'},explode=explosion,shadow=True)
plt.savefig('my_pie_plot')
plt.show()
```

```python
## Boxplot()
plt.boxplot(bullet['price'])    ## number summary
plt.show()
```

```python
## number summary= min,max,median,q=qartile(q1=25%,q2=50%,q3=75%), circles=outliers for finding this we use the boxplot
# index=(PER/100)*(N+1)
```

```python
bullet_filter_data=bullet[(bullet['price']>=5000) & (bullet['price']<=200000)]
```

```python
plt.boxplot(bullet_filter_data['price'])
plt.show()
```
